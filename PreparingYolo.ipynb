```python
import cv2
import numpy as np
import copy
from matplotlib import pyplot as plt
```

```python
imageName = "Psic_005Ev2.pdf_pg-16.jpg"
img1 = cv2.imread(imageName)
height, width = img1.shape[:2]

BB = [72, 267, 381, 362]

bb_xSuperiorIzq = BB[0]
bb_ySuperiorIzq = BB[1]

bb_h = BB[2]
bb_w = BB[3]

print("bb_xSuperiorIzq: {}, bb_ySuperiorIzq: {}".format(bb_xSuperiorIzq,bb_ySuperiorIzq))

cv2.circle(img1,(bb_xSuperiorIzq,bb_ySuperiorIzq), 10, (0,0,255), -1)

#point inferior derecho
bb_xInferiorDerecho = BB[0] + bb_w
bb_yInferiorDerecho = BB[1] + bb_h

print("bb_xInferiorDerecho: {}, bb_yInferiorDerecho: {}".format(bb_xInferiorDerecho,bb_yInferiorDerecho))

cv2.circle(img1,(bb_xInferiorDerecho,bb_yInferiorDerecho), 10, (0,0,255), -1)
plt.imshow(img1)
plt.show()

#print rectangles
cv2.rectangle(img1,(bb_xSuperiorIzq,bb_ySuperiorIzq),(bb_xInferiorDerecho,bb_yInferiorDerecho),(0,255,0),3)
plt.imshow(img1)
plt.show()

#center point of bb
x_center = round((bb_xInferiorDerecho + bb_xSuperiorIzq) / 2)
y_center = round((bb_ySuperiorIzq + bb_yInferiorDerecho) / 2)
print("x_center: {}, y_center: {}".format(x_center,y_center))

cv2.circle(img1,(x_center,y_center), 10, (0,0,255), -1)
plt.imshow(img1)
plt.show()
```

```python
#Normalizar valores para que estén entre 0 y 1
xYolo = round(x_center / width, 6) 
yYOLO = round(y_center / height, 6) 
wYOLO = round(bb_w / width, 6) 
hYOLO = round(bb_h / height, 6) 

print("xYolo: {}, yYOLO: {}".format(xYolo,yYOLO))
print("wYOLO: {}, hYOLO: {}".format(wYOLO,hYOLO))
```
